```python
###KAMERADAN EL

import cv2
import mediapipe as mp

cap = cv2.VideoCapture(0)
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

shrtTxt = "Bu tur insanlar cogunlukla cekici, etkileyici ve konuskan bir yapiya sahiptirler"
lngTxt = "Isaret parmagi, yuzuk parmagindan buyuk insanlar genellikle kendilerine guvenen insanlardir."
eqTxt = "Oldukca sakin, iyi huylu ve fikir ayriliklarindan hoslanmayan bir yapiya sahipsiniz."


while True:
    ret, image = cap.read()
    x , y, c = image.shape
    imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(imageRGB)
    
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                h, w, c = image.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                if id == 8 :
                    indexFinx = cx
                    indexFiny = cy
                    mpDraw.draw_landmarks(image, handLms, mpHands.HAND_CONNECTIONS)
                if id == 16 :
                    ringFinx = cx
                    ringFiny = cy
                    mpDraw.draw_landmarks(image, handLms, mpHands.HAND_CONNECTIONS)
        cv2.line(image, (indexFinx, indexFiny) , (ringFinx, ringFiny), (0,255,0), 2) 
        if ( indexFiny < ringFiny):
            cv2.putText(image, lngTxt, (0,x-100), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)
        elif (indexFiny == ringFiny):
            cv2.putText(image, eqTxt, (0,x-100), cv2.FONT_HERSHEY_SIMPLEX,  0.5, (255,255,255), 1)
        else:
            cv2.putText(image, shrtTxt, (0,x-100), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)

            
    cv2.imshow("Output", image)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
